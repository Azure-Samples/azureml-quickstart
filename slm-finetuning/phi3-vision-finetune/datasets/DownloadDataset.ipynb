```python
import os
import base64
from io import BytesIO
from PIL import Image
from datasets import load_dataset
import pandas as pd
```

```python
# Decode and save the images
def decode_and_save_image(row):
    doc_id = row['doc_id']
    encoded_image = row['encoded_image']
    
    # Create the folder if it doesn't exist
    os.makedirs('images', exist_ok=True)
    
    # Decode the base64 image
    decoded_image = base64.b64decode(encoded_image)
    
    # Save the image
    image_path = os.path.join('images', doc_id)
    with open(image_path, 'wb') as f:
        f.write(decoded_image)
    
    # Open the image using PIL and save it as JPEG
    with Image.open(BytesIO(decoded_image)) as img:
        img.save(image_path, 'JPEG')

df.apply(decode_and_save_image, axis=1)

# Display the DataFrame
df = df[['doc_id', 'content', 'document_type', 'key_details', 'insights']]
df
```

```python
dataset_dir = "./Sujet-Finance-Vision-10k"
dataset_path = os.path.join(dataset_dir, 'sujet_dataset.csv')
df.to_csv(dataset_path, index=False)
```
